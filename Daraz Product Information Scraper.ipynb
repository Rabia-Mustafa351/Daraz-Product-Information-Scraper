```python
options=Options()
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://www.daraz.pk/#') 
time.sleep(7)
driver.find_element_by_xpath('//*[@id="hp-flash-sale"]/div[2]/div[1]/a').click()

names=[]
orignalprice=[]
stars=[]
discountedPrice=[]
disPercentage=[]
ratings=[]
revew2=[]
revew1=[]
revew3=[]
st=1
st2=""
for i in range(25):
    st2=str(st)
    time.sleep(5)
    for m in driver.find_elements_by_xpath('//*[@id="campaign-1637348400000"]/div/a['+st2+']/div/div[2]/div[1]'):
        time.sleep(7)
        m.click()
        for m in driver.find_elements_by_class_name('pdp-mod-product-badge-wrapper'): # name
            k=m.find_element_by_class_name('pdp-mod-product-badge-title').text
            names.append(k)
            print(k)
        for m in driver.find_elements_by_class_name('score'):    # stars
            k=m.find_element_by_class_name('score-average').text
            stars.append(k)
            print(k)

        for m in driver.find_elements_by_class_name('origin-block'):   # discount  percentage
            k=m.find_element_by_class_name('pdp-product-price__discount').text
            print(k)
            disPercentage.append(k)
        k=driver.find_element_by_xpath('//*[@id="module_product_price_1"]/div/div/span').text #discounted price
        print(k)
        discountedPrice.append(k)
        k=driver.find_element_by_xpath('//*[@id="module_product_price_1"]/div/div/div/span[1]').text #original price
        print(k)
        orignalprice.append(k)
        #for m in driver.find_elements_by_class_name('pdp-product-price'): #discounted price
            #k=m.find_element_by_class_name('pdp-price pdp-price_type_normal pdp-price_color_orange pdp-price_size_xl').text
            #print(k) 
        #for m in driver.find_elements_by_class_name('pdp-mod-product-price'):  # discounted price
            #k=m.find_element_by_class_name(' pdp-price pdp-price_type_normal pdp-price_color_orange pdp-price_size_xl').text
            #print(k)
        for m in driver.find_elements_by_class_name('container-star pdp-review-summary__stars pdp-stars_size_s'):
            k=m.find_element_by_class_name('pdp-link pdp-link_size_s pdp-link_theme_blue pdp-review-summary__link').text
            print(k)
        for m in driver.find_elements_by_class_name('summary'):
            k=m.find_element_by_class_name('count').text
            print(k)
            ratings.append(k)
        k=driver.find_element_by_xpath('//*[@id="module_product_review"]/div/div[3]/div[1]/div[1]/div[3]/div[1]').text
        print(k)
        revew1.append(k)
        k=driver.find_element_by_xpath('//*[@id="module_product_review"]/div/div[3]/div[1]/div[2]/div[3]/div[1]').text
        print(k)
        revew2.append(k)
        k=driver.find_element_by_xpath('//*[@id="module_product_review"]/div/div[3]/div[1]/div[3]/div[3]/div[1]').text
        print(k)
        revew3.append(k)
        driver.back()
        time.sleep(5)
        st=st+1
        print(st)
```
